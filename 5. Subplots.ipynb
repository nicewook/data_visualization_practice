## Subplotting
Subplotting 은 여러개의 plot을 한꺼번에 그리는거다. 
subplot 에는 행과 열만 지정해주면 된다. (raws, columns)

```python
import matplotlib.pyplot as plt
fig, axarr = plt.subplots(2, 1, figsize=(12,8))
```

```python
#fig 에는 전체 plot 이 리턴된다. 
print(fig)

#axarr 는 axis 의 array 정도가 되려나? 각각의 plot 들이 AxesSubplot 이라는 object 로서 리턴된다 
print(axarr)
```

```python
# 위, 아래중 어디다 그릴지는 어떻게  명시할까? 

fig, axarr = plt.subplots(2, 1, figsize=(12,8))

# 위쪽에 그려보자 = axarr[0]
reviews['points'].value_counts().sort_index().plot.bar( ax=axarr[0] )

# 아래쪽에 그려보자 = axarr[1]
reviews['province'].value_counts().head(20).plot.bar( ax=axarr[1] )
```

```python
# 마구 만들어보자

fig, axarr = plt.subplots(6, 2, figsize=(12,8))


# 아무거나 때려넣자. 아까는 1차원 배열이지만 이번엔 2차원 배열임에 주의

reviews['points'].value_counts().sort_index().plot.bar( ax=axarr[0][0] )
reviews['province'].value_counts().head(20).plot.bar( ax=axarr[0][1] )
reviews['points'].value_counts().sort_index().plot.bar( ax=axarr[1][1] )
reviews['province'].value_counts().head(20).plot.bar( ax=axarr[1][0] )
reviews['points'].value_counts().sort_index().plot.bar( ax=axarr[2][0] )
reviews['province'].value_counts().head(20).plot.bar( ax=axarr[2][1] )
reviews['points'].value_counts().sort_index().plot.bar( ax=axarr[3][1] )
reviews['province'].value_counts().head(20).plot.bar( ax=axarr[3][0] )
reviews['points'].value_counts().sort_index().plot.bar( ax=axarr[4][0] )
reviews['province'].value_counts().head(20).plot.bar( ax=axarr[4][1] )
reviews['points'].value_counts().sort_index().plot.bar( ax=axarr[5][1] )
reviews['province'].value_counts().head(20).plot.bar( ax=axarr[5][0] )
```

```python
#  2차원 배열임을 확인해보자

print(axarr)
```

```python
# font와 color 로 먹여보고
# sns.despine() 박스도 없애주고
# hspace 옵션도 조절해보자 

fig, axarr = plt.subplots(2, 2, figsize=(12,8))

reviews['points'].value_counts().sort_index().plot.bar( ax=axarr[0][0], fontsize=12, color='red' )
reviews['province'].value_counts().head(20).plot.bar( ax=axarr[0][1], fontsize=12, color='blue'  )
reviews['points'].value_counts().sort_index().plot.bar( ax=axarr[1][1], fontsize=12, color='green'  )
reviews['province'].value_counts().head(20).plot.bar( ax=axarr[1][0], fontsize=12, color='violet'  )

axarr[0][0].set_title("RED",  fontsize=18)
axarr[0][1].set_title("BLUE",  fontsize=18)
axarr[1][0].set_title("VIOLET",  fontsize=18)
axarr[1][1].set_title("GREEN",  fontsize=18)

#hspace 는 각 subplot 간의 세로 간격을 말하는 듯 하다
plt.subplots_adjust(hspace=2)

import seaborn as sns
sns.despine()

```

### 퀴즈와 답변

퀴즈

    A matplotlib plot consists of a single X composed of one or more Y. What are X and Y?
    The subplots function takes which two parameters as input?
    The subplots function returns what two variables?

답변

    matplotlib의 plot은 하나의 figure 와 하나이상의 axis 로 이루어져 있다. 
    subplots 는 행과 열이라는 두 파라미터를 가진다.
    subplots 는 하나의 figure 값과, 여러개의 axis 로 이루어진 배열을 리턴해준다. 
    
원문 답변

    The plot consists of one overall figure composed of one or more axes.
    The subplots function takes the number of rows as the first parameter, and the number of columns as the second.
    The subplots function returns a figure and an array of axes.



```python
# 실전 디자인 스킬 향상
# 일단은 도표를 하나 가져오자

import pandas as pd
import matplotlib.pyplot as plt
pokemon = pd.read_csv("./data/Pokemon_with_stat.csv")
pokemon.head()
```

```python
# 일단 두 개의 subplot 을 만들자
fig, axarr = plt.subplots(2, 1, figsize=(8,8))
```

```python
# 각각 attack rating, defence rating을 넣어주자. 
# 좀 이쁘게 꾸며주기도 하자
fig, axarr = plt.subplots(2, 1, figsize=(8,8))

pokemon['Attack'].plot.hist( ax=axarr[0])
pokemon['Defense'].plot.hist( ax=axarr[1])

axarr[0].set_title("Pokemon Attack Ratings", fontsize=18)
axarr[1].set_title("Pokemon Defense Ratings", fontsize=18)
```
