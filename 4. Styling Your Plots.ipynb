```python
import pandas as pd
reviews = pd.read_csv("./data/winemag-data_first150k.csv", index_col=0)
reviews.head(10)
```

```python
# bar 형태로 만드는 것 기억나지?
reviews['points'].value_counts().sort_index().plot.bar()
```

```python
# figsize 를 넣어주면, 인치 단위로 가로세로를 정할 수 있다.
reviews['points'].value_counts().sort_index().plot.bar(figsize = (13,7))
```

```python
# 색깔 통일!
reviews['points'].value_counts().sort_index().plot.bar(color='mediumvioletred')
```

```python
# 글자 크기
reviews['points'].value_counts().sort_index().plot.bar(fontsize=16)
```

```python
# 타이틀도 넣자
reviews['points'].value_counts().sort_index().plot.bar(title="Wine Points")
```

```python
# 타이틀을 키워주려면 matplotlib 의 도움이 필요하다
import matplotlib.pyplot as plt

ax = reviews['points'].value_counts().sort_index().plot.bar(title="Wine Points")
ax.set_title("Wine Points 2017", fontsize=20)
```

```python
#Seaborn 도 써보자
import matplotlib.pyplot as plt
import seaborn as sns

ax = reviews['points'].value_counts().sort_index().plot.bar()
ax.set_title("Wine Points 2017", fontsize=30)
sns.despine(bottom=True, left=True)
```

# 실습해보자

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pokemon = pd.read_csv("./data/Pokemon_with_stat.csv")
pd.set_option("Max_columns",None)
pokemon.head(5)
```

```python
# Scatter 를 해보자 attack 과 defences 를 가지고 
pokemon.plot.scatter(x='Attack', y='Defense', figsize=(12,6), title="Pokemon by Attack and Defense")
```

```python
# Scatter 를 해보자 attack 과 defences 를 가지고 
ax = pokemon.plot.scatter(x='Attack', y='Defense', figsize=(12,6), title="Pokemon by Attack and Defense")
ax.set_title("Pokemon", fontsize=20)
```

```python
#Pokemon by Stat Total
ax = pokemon['Total'].plot.hist(
    figsize=(12, 6),
    fontsize=14,
    bins=50,
    color='gray'
)
ax.set_title('Pokemon by Stat Total', fontsize=20)
```

```python
#Pokemon by Primary Type
ax = pokemon['Type 1'].value_counts().plot.bar(figsize=(12,6))
sns.despine(left=True, bottom=True)
```
