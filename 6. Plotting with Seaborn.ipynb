Seaborn 은 standalone 데이터 시각화 패키지이며, 이 하나의 패키지가 음청 끝내주는 시각화를 제공합니다.

```python
# 가지고 놀 재료를 준비합니다.

import pandas as pd
import seaborn as sns

reviews = pd.read_csv("./data/winemag-data_first150k.csv", index_col=0)
```

```python
# pandas 에서 쓰던 plot.bar() 대신
# seaborn 의 countplot() 을 써봅시다

sns.countplot(reviews['points'])  
```

```python
# 포인트는 pandas 였다면 .value_counts().sort_index() 붙이고 난리났을 거라는 것

reviews['points'].value_counts().sort_index().plot.bar()
```

```python
# 선으로 그리는 건 조금 복잡하다
# KDE = Kernel Density Estimate - 데이터의 노이즈를 줄여주는 통계학적 기법이다.
# 따라서 전체적인 추세를 보기 좋다. 아래 판다스의 출렁이는 plot 과 비교해보라

sns.kdeplot(reviews.query('price<200').price)
```

```python
# 판다스였다면? 똑같이는 못그리겠고 비슷하게 그리자면.

reviews[reviews['price'] < 200]['price'].value_counts().sort_index().plot.line()
```

```python
# KED - Kernel Density Estimate 를 2차원에서 써보자
# 일단 deprecated 된 기능이라고는 하지만 일단 돌아간다. 
# 이건 연산이 빡세다는 건 참고할 것. 그래서 5000개만 샘플링 한거다

sns.kdeplot(reviews[reviews['price']<200].loc[:, ['price', 'points']].dropna().sample(5000))
```

```python
# 판다스 히스토그램 꺼져! distplot 을 쓰겠다.
sns.distplot(reviews['points'], bins=10, kde=False)
```

```python
# 판다스 히스토그램 꺼져! distplot 을 쓰겠다2
# bins 는 히스토그램을 몇등분할지를 말하는 거구나

sns.distplot(reviews['points'], bins=20, kde=False)
```

```python
# 판다스 히스토그램 꺼져! distplot 을 쓰겠다3
# kde 를 먹여주면 선을 겹쳐서 보여주는 구만

sns.distplot(reviews['points'], bins=20, kde=True)
```

```python
# 판다스는 이렇게 했었지?
reviews['points'].plot.hist()
```

```python
# Scatter 해줄때는 jointplot 을 해준다
# bells and whistles - 위와 오른쪽에 히스토그램도 보여주고 pearsonr, p 같은 값도 보여주고 별거별거 장식적일 수 있는거 많이 붙어있음

sns.jointplot(x='price', y='points', data=reviews[reviews['price'] < 200])
```

```python
# 여기다가 kind='hex', grid_size='20' 만 넣어줘보자

sns.jointplot(x='price', y='points', data=reviews[reviews['price'] < 200], kind='hex', gridsize=20)
```

```python
# Boxplot 은 통계학적으로 매우 유용하다. 
# 박스 안이 전체 값의 25-75 (=50) 안에 드는 값들을 담고 있다는 것을 의미한다. = 박스가 납작하다면 거기에 많이 몰려있다는 것
# 2,3,4번째의 점들은 이상한 넘들로 통계에 넣기 그런 놈들 = outliar
# 여기선 데이터에 대한 가정이 하나 있다. 정규분포일 거라는 가정


df = reviews[reviews.variety.isin(reviews.variety.value_counts().head(5).index)]  # 뭔 소리인지는 모르겠지만

sns.boxplot(
    x='variety', 
    y='points',
    data=df
)
```

```python
# violinplot 을 보자 - 바이올린을 닮아서 그런가보다
# 옆으로 울퉁불퉁한건 그 속에서의 분포를 나타내는 거구만 (kernel density estimate 이다)
# 이게 boxplot 보다는 좀더 많은걸 보여준다 싶다 

sns.violinplot(
    x='variety',
    y='points',
    data=reviews[reviews.variety.isin(reviews.variety.value_counts()[:5].index)]
)
```

## 질문과 답변
    
    A seaborn countplot is equivalent to what in pandas?
    A seaborn jointplot is equivalent to a what in pandas?
    Why might a kdeplot not work very well for ordinal categorical data?
    What does the "box" in a boxplot represent?

## 나의 답변
    countplot 은 plot.bar
    jointplot 은 scatter
    kedplot 은 부드럽게 경향을 보여준다. 그러한 경향이 없는 데이터에는 문제가 될 수 있다.
    boxplot 의 box 는 전체 분포의 25%-75% 값들의 위치를 의미한다. 그 속에서도 25%-50%, 50%-75% 를 박스중간을 잘라 보여준다
    
## 정답
    

    A seaborn countplot is like a pandas bar plot.
    A seaborn jointplot is like a pandas hex plot.
    KDEPlots work by aggregating data into a smooth curve. This is great for interval data but doesn't always work quite as well for ordinal categorical data.
    The top of the box is the 75th percentile. The bottom of the box is the 25th percentile. The median, the 50th percentile, is the line in the center of the box. So 75% of the data in the distribution is located within the box!


    


```python
## 실습도 해보자. 재료를 준비하자

pokemon = pd.read_csv("./data/Pokemon_with_stat.csv")
pokemon.head()
```

```python
# Generation 별로 바로 보여주자
# bar 는 countplot 이었지?

sns.countplot(pokemon['Generation'])
```

```python
# 이번에는 히스토그램이다.
# 히스토그램의 촘촘함음 bins 이고  kde를 True 로 해주면 선도 그어줬지?

sns.distplot(pokemon['HP'], bins=50, kde=True)


```

```python
# 이번에는 장식이 많은 녀석 (bells and whistles)
# jointplot 이다

sns.jointplot(x='Attack', y='Defense', data=pokemon)
```

```python
# jointplot 을 hex 처럼 

sns.jointplot(x='Attack', y='Defense', data=pokemon, kind='hex', gridsize=20)
```

```python
# kdeplot 을 2차원에 적용해보자

sns.kdeplot(pokemon[['HP','Attack']])
```

```python
# boxplot

sns.boxplot(x="Legendary", y="Attack", data=pokemon)
```

```python
# violneplot

sns.violinplot(x="Legendary", y="Attack", data=pokemon)
```
