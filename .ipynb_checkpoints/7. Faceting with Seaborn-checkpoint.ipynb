# 7. Faceting with Seaborn
- 링크: https://www.kaggle.com/residentmario/faceting-with-seaborn/
- facet 은 "한 면"이라는 말인데 이 제목은 무슨 뜻일까? 
- 그냥 seaborn 의 subplot 과 비슷하다 생각해도 될거 같다. 
- Univariate 는 변수 하나, bivariate 는 변수 두개라 생각해버리자

```python
# 이건 데이터 전처리 부분인데 복잡해서 이해하기 어려우면 그냥 넘어가자

import pandas as pd
pd.set_option('max_columns', None)
df = pd.read_csv("./data/CompleteDataset.csv", index_col=0)
df.head()
```

```python
import re
import numpy as np

footballers = df.copy()  # df 원본은 안건드리고 복사해서 씀

# 'Value' 는 €95.5M 이런 식인데 마지막 M 만 떼어서 'Unit'에 넣겠다는 건가 보다
footballers['Unit']  = df['Value'].str[-1]  
print(footballers['Unit'])

# 이건 새로운 필드를 만들고, Value 의 뒤에 M, K 등의 값이 없는 녀석들만 다 날려주라는 것으로 보임
footballers['Value (M)'] = np.where(footballers['Unit'] == '0', 0, 
                                    footballers['Value'].str[1:-1].replace(r'[a-zA-Z]',''))
print(footballers['Value (M)'])  

# 실수로 캐스팅 해주고 1/1000 로 값 줄이기 
footballers['Value (M)'] = footballers['Value (M)'].astype(float)
footballers['Value (M)'] = np.where(footballers['Unit'] == 'M', 
                                    footballers['Value (M)'], 
                                    footballers['Value (M)']/1000)

# 축구선수의 연봉과 선호 포지션 관계를 보려는 건가?
footballers = footballers.assign(Value=footballers['Value (M)'],
                                 Position=footballers['Preferred Positions'].str.split().str[0])
```

```python
footballers.head()
```

```python
import seaborn as sns
```

```python
# FacetGrid
# 이 모든 축구선수 데이터를 스트라이커와 골키퍼로 나눠서 보고 싶다. 

df = footballers[footballers["Position"].isin(['ST', 'GK'])]
g = sns.FacetGrid(df, col="Position")
```

```python
# 여기까지는 아무 의미가 없구나. map을 먹여줘야 하네.
# 즉, g 까지 만들어준 다음에는  그냥 sns 먹이는 것과 비슷하게 해주면 둘로 나뉘어진대로 각각 연산됨
df = footballers[footballers["Position"].isin(['ST', 'GK'])]
g = sns.FacetGrid(df, col="Position")
g.map(sns.kdeplot, "Overall")
```

```python
# 이번에는 두 개 말고 다 보자
df = footballers

g = sns.FacetGrid(df, col="Position", col_wrap=4)
g.map(sns.kdeplot, "Overall")
```

```python
# 먼저 포지션에서 ST, GK 만 걸러낸다음
# 다시 Club 에서 세 팀만 걸러낸다

df = footballers[footballers['Position'].isin(['ST', 'GK'])]  
df = df[df['Club'].isin(['Real Madrid CF', 'FC Barcelona', 'Atlético Madrid'])]
df.head()

```

```python
# 먼저 포지션에서 ST, GK 만 걸러낸다음
# 다시 Club 에서 세 팀만 걸러낸다

df = footballers[footballers['Position'].isin(['ST', 'GK'])]  
df = df[df['Club'].isin(['Real Madrid CF', 'FC Barcelona', 'Atlético Madrid'])]

# 그리고 그러한 df 를 row, col 에 각각 Position 과 Club 으로 나누어 FacetGrid 를 먹여준다
# 1행, 2행은 각각 Position, 1,2,3 열은 각각 Club 이 된다. 
g = sns.FacetGrid(df, row="Position", col="Club")
g.map(sns.violinplot, "Overall")
```

```python
# 행과 열의 순서를 임의로 (=arbitrarily) 하지 않고 명시를 할 수 도 있다.

df = footballers[footballers['Position'].isin(["ST", "GK"])]
df = df[df["Club"].isin(['Real Madrid CF', 'FC Barcelona', 'Atlético Madrid'])]

g = sns.FacetGrid(df, row="Position", col="Club",
                  row_order=["GK", "ST"],
                  col_order=['Atlético Madrid', 'FC Barcelona', 'Real Madrid CF'])
g.map(sns.violinplot, "Overall")
```

- Facet 은 가로와 세로에 각각의 변수를 주고 사용하기에 최적이다. 
- 단점은 너무 많은 다차원이나, 변수의 항목개수가 많은 경우에는 사용하는데 문제가 있다는 것이다.


# Pairplot 을 보자 

```python
# 바로 코드를 보는게 빠르다

sns.pairplot(footballers[['Overall', 'Potential', 'Value']])
```

    You should try to keep your grid variables down to five or so. Otherwise the plots get too small.
    It's (1) a multivariate technique which (2) is very easy to use.
    Pair plots are most useful when just starting out with a dataset, because they help contextualize relationships within it.



# 실습도 해보자


```python
import pandas as pd
import seaborn as sns

pokemon = pd.read_csv("./data/Pokemon_with_stat.csv", index_col=0)
pokemon.head()
```

```python
# Legendary 별로 subplot 들을 그려보자
g = sns.FacetGrid(pokemon, row="Legendary")
g.map(sns.kdeplot, "Attack")  # Kernel Density Estimate
```

```python
# 이번에는 행은 Generation 이고 열은 Lengendary 로 나눠보자
#  이 조합별로 Attack 값을 보려는 것이다. 

g = sns.FacetGrid(pokemon, col="Legendary", row="Generation")
g.map(sns.kdeplot, "Attack")
g.map(sns.violinplot, "Attack")
```

```python
# 이번에는 pairplot

sns.pairplot(pokemon[['HP', 'Attack', 'Defense']])  # 세가지 조합으로 그려보는 것 같다 
```
